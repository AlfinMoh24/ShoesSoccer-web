```python
import pandas as pd
import pymysql
import ast
```

```python

def get_all_products_from_database():
    # Koneksi ke database MySQL
    conn = pymysql.connect(host='localhost', user='root', password='', db='shoes-soccer')
    cursor = conn.cursor(pymysql.cursors.DictCursor)
    
    # Query untuk mengambil semua data produk beserta vektor-fiturnya
    query = """
     SELECT sepatu.*, 
            brands.brand_name, 
            surfaces.description as surface_description, 
            surfaces.surface_type,
            materials.material, 
            series.description as series_description,
            series.fungsi as series_fungsi,
            materials.description as material_description,
            series.series, 
            models.model, 
            positions.position 
        FROM sepatu
        LEFT JOIN brands ON sepatu.brand_id = brands.brand_id
        LEFT JOIN surfaces ON sepatu.surface_id = surfaces.surface_id
        LEFT JOIN materials ON sepatu.material_id = materials.material_id
        LEFT JOIN series ON sepatu.series_id = series.series_id
        LEFT JOIN models ON sepatu.model_id = models.model_id
        LEFT JOIN positions ON sepatu.position_id = positions.position_id
    """
    
    cursor.execute(query)
    products = cursor.fetchall()
    conn.close()
    return products
```

```python
sepatu = get_all_products_from_database()
df = pd.DataFrame(sepatu)
```

```python
for column in ['color_vector','brand_vector', 'surface_vector', 'material_vector', 'position_vector', 'series_vector']:
    df[column] = df[column].apply(lambda x: ast.literal_eval(x.decode('utf-8')))
```

```python
data_sepatu_with_vector = df[['shoes_id','kode_produk', 'shoes_name', 'color','price', 'size', 'brand_name','surface_type','surface_description','material','series','series_description','series_fungsi','position','image_url','color_vector','brand_vector','surface_vector','material_vector','position_vector','series_vector']]
```
